## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

class Lane(object):
    
    def __init__(self): 
        self.line_coeffs = None
    
    def update_lane(self, segments):
        if not segments:
            # do not draw if we did not detect any lines
            self.line_coeffs = None
            return

        self.line_coeffs = fit_line(segments)
    
    def has_line_coeffs(self):
        return self.line_coeffs is not None
    
    def get_line_coeffs(self):
        return self.line_coeffs
        
class Lanes(object):
    
    left_lane = Lane()
    right_lane = Lane()
    
    def update_lanes(self, lines):
        
        left_lane_segments = []
        right_lane_segments = []
    
        for line in lines:
            for x1,y1,x2,y2 in line:
                m = ((y2-y1)/(x2-x1))
                if m > 0:
                    left_lane_segments += [line]
                else:
                    right_lane_segments += [line]

        self.left_lane.update_lane(left_lane_segments)
        self.right_lane.update_lane(right_lane_segments)
    
    def has_lanes(self):
        return self.left_lane.has_line_coeffs() and self.right_lane.has_line_coeffs()
    
    def get_intersection_point(self):
        equation = self.left_lane.get_line_coeffs() - self.right_lane.get_line_coeffs()
        x = -equation[1] / equation[0]
        y = np.poly1d(self.left_lane.get_line_coeffs())(x)
        x, y = map(int, [x, y])
        return [x, y]
    
    def get_left_lane(self):
        return self.left_lane
    
    def get_right_lane(self):
        return self.right_lane
        
def get_x_coord(z, y):
    return int((y - z[1]) / z[0])

def fit_line(segments):
    x = []
    y = []
    for line in segments:
        for x1, y1, x2, y2 in line:
            x += [x1, x2]
            y += [y1, y2]

    return np.polyfit(x, y, 1)

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def draw_lanes(img, lanes, color=[255, 0, 0], thickness=2):
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    if lanes.has_lanes():
        intersection_point = lanes.get_intersection_point()
        draw_lane(line_img, lanes.get_left_lane(), intersection_point, color, thickness)
        draw_lane(line_img, lanes.get_right_lane(), intersection_point, color, thickness)
    return line_img
    
def draw_lane(img, lane, intersection_point, color, thickness):
    intersection_offset = 20
    y1 = intersection_point[1] + intersection_offset
    y2 = img.shape[0] # image shape is [y, x]
    x1 = get_x_coord(lane.get_line_coeffs(), y1)
    x2 = get_x_coord(lane.get_line_coeffs(), y2)
        
    line_coords = [(x1, y1), (x2, y2)]
    cv2.line(img, line_coords[0], line_coords[1], color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns hough lines.
    """
    return cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )

def show_frame_by_index(clip, frame): 
    for idx, val in enumerate(clip.iter_frames()):
        if idx == frame:
            #plt.imshow(process_image(val))
            plt.imshow(val)
            break
````

Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

````python
import functools as func

def image_processor():
    return func.partial(process_image, lanes=Lanes())

def process_image(image, lanes):
    
    # 1) Process
    imshape = image.shape
    grey_img = grayscale(image)
    blurred_img = gaussian_blur(grey_img, 5)
    canny_img = canny(blurred_img, 50, 150)
    roi_vertices = np.array([[(50, 540), (420, 330), (590, 330), (image.shape[1] - 50, 540)]], dtype=np.int32)
    roi_canny_img = region_of_interest(canny_img, roi_vertices)
    lines = hough_lines(roi_canny_img, 2, math.pi / 180, 50, 40, 20)
    
    # 2) Update
    lanes.update_lanes(lines)
    
    # 3) Draw
    lines_img = draw_lanes(image, lanes, thickness=10)
    return weighted_img(lines_img, image)
    
````
